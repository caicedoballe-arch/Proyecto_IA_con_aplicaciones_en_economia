```python
#Librerias necesarias
import requests
import pandas as pd
from functools import reduce
```

```python
#Indicadores a utilizar
indicadores = {
    "FP.CPI.TOTL.ZG": "inflacion",
    "NY.GDP.MKTP.KD.ZG": "gdp_growth",
    "NY.GDP.PCAP.KD": "gdp_pc",
    "NY.GDP.PCAP.PP.KD": "gdp_pc_ppp",
    "NE.GDI.TOTL.ZS": "investment",
    "NV.IND.TOTL.ZS": "industry",
    "NV.AGR.TOTL.ZS": "agricultura",
    "SE.SEC.ENRR": "matricula_secundaria",
    "SE.TER.ENRR": "matricula_terciaria",
    "SP.DYN.LE00.IN": "life_expectancy",
    "SL.UEM.TOTL.ZS": "desempleo",
    "NE.EXP.GNFS.ZS": "exportaciones",
    "NE.TRD.GNFS.ZS": "comercio",
    "BX.KLT.DINV.WD.GD.ZS": "fdi",
    "IT.NET.USER.ZS": "internet",
    "SP.POP.GROW": "pop_growth"
}
```

```python
# Función para descargar un indicador
def descargar_indicador(indicador, nombre):
    url = f"https://api.worldbank.org/v2/country/all/indicator/{indicador}?format=json&per_page=20000"
    response = requests.get(url).json()
    data = response[1]
    df = pd.DataFrame(data)
    df['country_name'] = df['country'].apply(lambda x: x['value'] if isinstance(x, dict) else None)
    df = df[['countryiso3code', 'country_name', 'date', 'value']]
    df.columns = ['country_code', 'country_name', 'year', nombre]
    df['year'] = pd.to_numeric(df['year'], errors='coerce')
    return df

# Descarga de los indicadores
dfs = []
for codigo, nombre in indicadores.items():
    df_temp = descargar_indicador(codigo, nombre)
    dfs.append(df_temp)

# Union de los dataframes
df_final = reduce(
    lambda left, right: pd.merge(left, right, on=['country_code', 'country_name', 'year'], how='outer'),
    dfs
)

# Filtro de los años
df_final = df_final[(df_final['year'] >= 2000) & (df_final['year'] <= 2023)]

df_final.head()
```

```python
#Codigos de los paises
df_final = df_final[df_final['country_code'].str.len() == 3]
```
